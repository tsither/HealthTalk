# Formatting Spider dataset for finetuning with LLM

```python
from datasets import load_dataset
import pandas as pd


dataset = load_dataset("xlangai/spider")
```

```python
df_train = dataset['train'].to_pandas()

df_val = dataset['validation'].to_pandas()
```

```python
df_train.columns
```

```python
df_train.shape
df_val.shape
```

### Randomize data

```python
df_train = df_train.sample(frac=1)
df_val = df_val.sample(frac=1)
df_val
```

```python
fine_tuning_dataset = df_train
fine_tuning_dataset
```

[
  { "role": "system", "content": "You are a helpful assistant" },

```python
user_content_train = []
user_content_val = []


for idx, row in fine_tuning_dataset.iterrows():
    # print(row['query'])
    user_content_train.append((idx, row['question'], row['query']))


for idx, row in df_val.iterrows():
    # print(row['query'])
    user_content_val.append((idx, row['question'], row['query']))

    
```

```python
print(len(user_content_train))
```

```python
user_content_val[0]
```

```python
FT_LINES = []
VAL_LINES = []


for entry in user_content_train:
    # print(entry[1])
    j = {
  "messages": [
    { "role": "system", "content": "You are a senior data engineer. Given the user input, construct a valid, intelligent SQL query." },
    { "role": "user", "content": entry[1] },
    { "role": "assistant", "content": entry[2] },
  ]
}

    FT_LINES.append(j)


for entry_val in user_content_val:
    # print(entry_val[1])
    j_val = {
  "messages": [
    { "role": "system", "content": "You are a senior data engineer. Given the user input, construct a valid, intelligent SQL query." },
    { "role": "user", "content": entry_val[1] },
    { "role": "assistant", "content": entry_val[2] },
  ]
}

    VAL_LINES.append(j_val)
```

```python
VAL_LINES
```

```python
import json
```

```python
with open('PMA_FT_spider_TRAIN.jsonl', 'w') as file:
    # Iterate through the data and write each item as a JSON line
    for item in FT_LINES:
        json_line = json.dumps(item)
        file.write(json_line + '\n')

with open('PMA_spider_validation.jsonl', 'w') as file:
    # Iterate through the data and write each item as a JSON line
    for item in FT_LINES:
        json_line = json.dumps(item)
        file.write(json_line + '\n')
```
